## 13-1. BASEMODEL_DELAY

```python
import math
import pandas as pd
```

```python
basemodel_delay = pd.read_excel('C:/Users/user/1. collabo/final/2020.11.06/Airport_ALL_PRE_T.xlsx', index_col = 0)
```

```python
mae = 0
mse = 0


for i in range(len(basemodel_delay)):
    tmp = basemodel_delay['DELAY_SECOND7'][i]
    abs_tmp = abs(tmp)
    square_tmp = (tmp**2)
    mae += abs_tmp
    mse += square_tmp


print('mae : {}'.format(int(mae)/len(basemodel_delay)))
print('mse : {}'.format(int(mse)/len(basemodel_delay))) 
print('rmse : {}'.format(math.sqrt(int(mse)/len(basemodel_delay))))
```

## 13-2. BASEMODEL_ELAPSED

```python
basemodel_elapsed = pd.read_excel('C:/Users/user/1. collabo/final/2020.11.06/Airport_ALL_PRE_T.xlsx', index_col = 0)
```

```python
# 모델 예측 생성
basemodel_elapsed['BASEMODEL_PRE'] = basemodel_elapsed['ELASPED_SECOND4'] + (basemodel_elapsed['AVG_ELASPED_SECOND7'] - basemodel_elapsed['AVG_ELASPED_SECOND4'])
```

```python
# 모델 딜레이 생성
basemodel_elapsed['BASEMODEL_DELAY'] = basemodel_elapsed['BASEMODEL_PRE'] - basemodel_elapsed['ELASPED_SECOND7']
```

```python
basemodel_elapsed
```

```python
mae = 0
mse = 0


for i in range(len(basemodel_elapsed)):
    tmp = basemodel_elapsed['BASEMODEL_DELAY'][i]
    abs_tmp = abs(tmp)
    square_tmp = (tmp**2)
    mae += abs_tmp
    mse += square_tmp


print('mae : {}'.format(int(mae)/len(basemodel_elapsed)))
print('mse : {}'.format(int(mse)/len(basemodel_elapsed))) 
print('rmse : {}'.format(math.sqrt(int(mse)/len(basemodel_elapsed))))
```

## 13-3. BASEMODEL_DISTANCE

```python
import pandas as pd

from datetime import timedelta, datetime, time
from haversine import haversine
```

```python
basemodel_distance = pd.read_excel('C:/Users/user/1. collabo/final/2020.11.06/Airport_ALL_PRE.xlsx')
```

```python
basemodel_distance
```

```python
SID_list = basemodel_distance.drop_duplicates('SID', keep='first').SID.tolist()
```

```python
len(SID_list)
```

```python
#경도 고정
START = (37.57612, 126.6971) # (lat, lon)
BULTI = (36.7229, 126.6971)
BEDES = (36.1514, 126.6971)
MANGI = (35.5031, 126.6971)
DALSU = (35.1253, 126.6971)
DOTOL = (34.2543, 126.6971)
END = (33.48214, 126.6971)
```

```python
#전체거리
start_end = haversine(START, END)
print('start_end:', start_end)
print('')

#첫번째 안테나에서 도착지점까지의 거리
bulti_end = haversine(BULTI, END)
print('bulti_end:', bulti_end)
print('')

#두번째 안테나에서 도착지점까지의 거리
bedes_end = haversine(BEDES, END)
print('bedes_end:', bedes_end)
print('')

#세번째 안테나에서 도착지점까지의 거리
mangi_end = haversine(MANGI, END)
print('mangi_end:', mangi_end)
print('')

#네번째 안테나에서 도착지점까지의 거리
dalsu_end = haversine(DALSU, END)
print('dalsu_end:', dalsu_end)
print('')

#다섯번째 안테나에서 도착지점까지의 거리
dotol_end = haversine(DOTOL, END)
print('dotol_end:', dotol_end)
```

```python
import datetime
```

```python
#평균 소요시간 / 전체 거리
km_s = 2745.964504512/start_end
print('1km :', km_s, '초')
```

```python
start_end_s = start_end*km_s
print('start_end_s:', start_end_s)
print(str(datetime.timedelta(seconds = start_end_s)))
#54분 40초
print('')

bulti_end_s = bulti_end*km_s
print('bulti_end_s:', bulti_end_s)
print(str(datetime.timedelta(seconds = bulti_end_s)))
#43분 16초
print('')

bedes_end_s = bedes_end*km_s
print('bedes_end_s:', bedes_end_s)
print(str(datetime.timedelta(seconds = bedes_end_s)))
#35분 38초
print('')

mangi_end_s = mangi_end*km_s
print('mangi_end_s:', mangi_end_s)
print(str(datetime.timedelta(seconds = mangi_end_s)))
#26분 59초
print('')

dalsu_end_s = dalsu_end*km_s
print('dalsu_end_s:', dalsu_end_s)
print(str(datetime.timedelta(seconds = dalsu_end_s)))
#21분 56초
print('')

dotol_end_s = dotol_end*km_s
print('dotol_end_s:', dotol_end_s)
print(str(datetime.timedelta(seconds = dotol_end_s)))
#10분 18초
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_squared_log_error
```

```python
#시간 비례 거리
distance_list = []

for SID in SID_list:
    a = basemodel_distance[basemodel_distance.SID == SID].reset_index(drop = True)
    
    distance_list.append((a['ELASPED_SECOND'][3] + mangi_end_s))
```

```python
distance_list
```

```python
last_elasped_list = []

for SID in SID_list:
    a = basemodel_distance[basemodel_distance.SID == SID].reset_index(drop = True)
    
    last_elasped_list.append(a['ELASPED_SECOND'][6])
```

```python
last_elasped_list
```

```python
MAE =  mean_absolute_error(last_elasped_list, distance_list)
print('MAE : ', MAE)

MSE = mean_squared_error(last_elasped_list, distance_list)
print('MSE : ', MSE)

RMSE = np.sqrt(MSE)
print('RMSE : ', RMSE)

MSLE = mean_squared_log_error(last_elasped_list, distance_list)
print('MSLE : ', MSLE)

RMSLE = np.sqrt(mean_squared_log_error(last_elasped_list, distance_list))
print('RMSLE : ', RMSLE)

R2 = r2_score(last_elasped_list, distance_list)
print('R2 : ', R2)
```
