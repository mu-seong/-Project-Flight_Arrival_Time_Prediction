```python
import keras
import sklearn

from keras.wrappers.scikit_learn import KerasClassifier
from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization
from keras.callbacks import EarlyStopping, ModelCheckpoint

from sklearn.model_selection import train_test_split, GridSearchCV
```

<h1>데이터 정규화</h1>

```python
from sklearn.preprocessing import StandardScaler
standardscaler = StandardScaler()
```

```python
X_train.fillna(0)
```

```python
standardscaler.fit(X_train)
X_train = standardscaler.transform(X_train)
```

```python
X_test.fillna(0)
```

```python
print(standardscaler.fit(X_test))
X_test = standardscaler.transform(X_test)
```

```python
X_test
```

```python
y_train = np.asarray(y_train)
y_test = np.asarray(y_test)
```

<h1>모델링</h1>

```python
# y = airport_df[['ELASPED_SECOND7']] 일때

def build_model():
    model = tf.keras.models.Sequential()
    model.add(tf.keras.layers.Dense(64, activation = 'relu', input_shape = (X_train.shape[1], )))
    model.add(tf.keras.layers.Dropout(0.2))
    model.add(tf.keras.layers.Dense(64, activation = 'relu'))
    model.add(tf.keras.layers.Dropout(0.2))
    model.add(tf.keras.layers.Dense(64, activation = 'relu'))
    model.add(tf.keras.layers.Dropout(0.2))
    model.add(tf.keras.layers.Dense(64, activation = 'relu'))
    model.add(tf.keras.layers.Dropout(0.2))
    model.add(tf.keras.layers.Dense(64, activation = 'relu'))
    model.add(tf.keras.layers.Dense(1))
    
    #optimizer = tf.keras.optimizers.Nadam(lr=0.002, beta_1=0.9, beta_2=0.999, epsilon=None, schedule_decay=0.004)
    model.compile(loss = 'mse', optimizer = 'nadam', metrics = ['mae'])
    
    return model
```

```python
model = build_model()
model.summary()
```

<h1>모델 학습 및 평가</h1>

```python
from math import sqrt
```

```python
early_stopping = EarlyStopping(monitor = 'val_loss', patience = 100, verbose = 0)
mc = ModelCheckpoint('best_model.h5', monitor = 'val_loss', mode = 'min', save_best_only = True)
history = model.fit(X_train, y_train, epochs = 1000, batch_size = 64, validation_split = 0.2, callbacks = [early_stopping, mc])
```

```python
test_mse_score, test_mae_score = model.evaluate(X_test, y_test)
```

```python
print("MSE {}".format(test_mse_score))
print("RMSE {}".format(sqrt(test_mse_score)))
print("MAE {}".format(test_mae_score))
```
